```python
import pandas as pd
filePath ='DataFiles/Match.csv'
def readCsvData(filePath):
    return pd.read_csv(filePath)
df=readCsvData(filePath)
print(df)
```

```python
df.groupby('Season_Year')['Match_SK'].count()
```

### probelm statement
 
 1.2 Plot a bar graph  no. of matches won by each team in the year  2011
   - Description

```python
import matplotlib.pyplot as plt
import numpy as np
df2=df.copy()
sdf=pd.DataFrame(df2[df2['Season_Year'].isin(['2011'])])
wins=(sdf.groupby('match_winner')['Match_SK'].count())
teams=sdf.Team1.unique()
index=np.arange(len(teams))
plt.barh(teams,wins,color='purple')
```
